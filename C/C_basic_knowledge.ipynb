## C 语言简介

所有的 C 语言程序都需要包含 **main()**函数。 代码从 main() 函数开始执行。 

return 0； 表示退出程序 

**include** 预处理 将头文件插入到预处理命令的相应位置  

**<>**，表示系统自带的库 

“”用户自定义的库 
```C
通过命令行传参
main(int argc， char *argv[])
或者
main(int argc, char ** grgv) 
```
函数传参

**argc** 为:命令行参数个数(不给main()函数传递参数时默认值为1，即至少有一个参数为该可执行文件的文件名（含目录）)，

**argv** 是字符串数组，下标从0开始，第一个存放的是可执行程序的文件名字，然后一次存放传入的参数。 
(分别指向各个字符串参数的首地址，其中argv[0]存储的是可执行文件的文件名的首地址。) 

```C
#include <stdio.h>  
 
int main(int argc, char * argv) 
{  
    /* 我的第一个 C 程序 */  
    printf("可执行程序 %s ,参数个数为[%d], 运行输出：[%s]\n",argv[0], argc, argv[1]);  
    return 0; 
} 
```

字符串中‘\0’代表null 空字符，字符结束     

‘\n’代表换行 

此外在terminal 如果用 c 语言程序 一个空格代表一个输入 

NULL 表示空指针 空地址 

‘\0’  表示空字符 

C 语言是为了编写UNIX操作系统而诞生的，所产生的代码运行和读写速度和汇编语言几乎一样


## 2.C 环境设置

C语言运行和编写 需要文本编辑器和编译器 

**文本编辑器** 

通过编辑器创建的文件通常称为源文件，源文件包含程序源代码。 

C 程序的源文件通常使用扩展名 .c。 

常用的编辑器有 vscode， sublime text, atom, vim.emacs,eclipse 

**C编译器**

把源代码编译为可执行程序 

GNU 的c/C++编译器 
```sh
Gcc –v #查看是否有gcc 
```

gcc 有一些常用编译选项 

42 中要求启用 -Wall -Wextra –Werror 编译选项 

**1 -Wall** 

这是一个非常常用的编译选项，用于启用一批比较常见且易于修改的警告，这些选项都是对代码进行基本的检查，比如下面这些 

**2 -Wextra** 

单单只有-Wall可能还不够严格，GCC还有-Wextra作为补充，包括另外一些没有被-Wall包含的警告类型，比如

| 选项 | 作用 |
| ---- | ---- | 
|-Wcast-function-type|当函数被强转为不兼容的函数指针时发出警告|
| -Wempty-body | 当存在空的if、else或者do while语句时发出警告 |
|-Wunused-parameter |当函数有未被使用的参数时发出警告，需配合-Wall |
|-Wunused-but-set-parameter |当存在被设置但是未被使用的参数发出警告，需配合-Wall |
|-Wsign-compare |当比较有符号和无符号值时发出警告 |

**3 -Werror**

-Werror用于将所有警告视为错误。 

**4 -Wpedantic** 

对于所有不符合ISO C/ISO C++语言标准的源代码发出警告，等价于-pedantic。 

-pedantic-errors参数将这些警告视为错误，等同于-Werror=pedantic 

**5 -Wshadow** 

当局部变量屏蔽（shadow）已有已有变量时发出警告

## 3.程序结构

主要包括：
```shell
--预处理指令 
--函数 
--变量 
--语句&表达式 
--注释 
```

**预处理指令**

程序的第一行 ``#include <stdio.h>`` 是预处理器指令，告诉 C 编译器在实际编译之前要包含 stdio.h 文件。
除此之外 #define #ifdefien #idndefne 等都是预处理指令
